# Nominal/ OHE Encoding

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({'color':['red','green','blue','red','green','blue','red','green','blue']})
```

```python
df
```

```python
# Create an instance of OHE
encoder = OneHotEncoder()
```

```python
# Alphabetically sort and arrange data into matrix
# >> fitting the encoder to dataframe and transforming the categorical variable
# encoder.fit_transform(df[['color']]).toarray()
encoded = encoder.fit_transform(df[['color']])
```

```python
import pandas as pd
encoded_df= pd.DataFrame(encoded.toarray(),columns=encoder.get_feature_names_out())
```

```python
encoder.get_feature_names_out()
```

```python
encoded_df
```

```python
pd.concat([df,encoded_df],axis=1)
```

# OHE for Smoker:
